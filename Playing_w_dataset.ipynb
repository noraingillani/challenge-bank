```python
df = pd.read_csv("/content/titanic.csv")
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
print("Summary Statistics: ")
print(df.describe())
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].median(),inplace=True)
df['Embarked'].fillna(df['Embarked'].mode(0),inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df['Cabin'].fillna(df['Cabin'].mode()[0],inplace=True)
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
df.head()
```

```python
df.drop(['PassengerId', 'Ticket', 'Cabin', 'Name'], axis=1, inplace=True)
```

```python
df.head()
```

```python
first_class_passengers = df[df['Pclass'] == 1]
print("Number of first-class passengers:", len(first_class_passengers))
```

```python
df = df[df['Fare'] <= 500]
```

```python
df.head()
```

```python
young_passengers=df.loc[df['Age']<18]
print("no. of passengers under 18:", len (young_passengers))
```

```python
df.loc[df['Age']<18,'Category'] = 'Child'
df.head(15)
```

```python
df['Sex'] =df['Sex'].map({'male': 0, 'female': 1})
df['Pclass'] = df['Pclass'].astype('category')
df.head()
```

```python
average_fare_by_class = df.groupby('Pclass')['Fare'].mean()
print("Average fare by class:\n", average_fare_by_class)
```

```python
plt.figure(figsize=(10, 6))
df['Age'].hist(bins=20, color='yellow', edgecolor='navy')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df['Age'].hist(bins=15, color='yellow', edgecolor='navy')
plt.title('Distribution of  Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
df['FamilType'] = pd.cut(df['FamilySize'], bins=[0, 1, 4, float('inf')], labels=['Single','Small','Large'])
df.head()
```

```python
numeric_df = df.select_dtypes(include=[np.number])
correlation_matrix = numeric_df.corr()
plt.figure(figsize=(12,10))
sns.heatmap(correlation_matrix,annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
